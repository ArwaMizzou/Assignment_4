```python
def read_graph(file_path):
    with open(file_path, 'r') as file:
        num_vertices = int(file.readline().strip())  # Read the number of vertices
        adjacency_list = {i: [] for i in range(1, num_vertices + 1)}  # Initialize adjacency list
        
        # Populate adjacency list with edges
        for line in file:
            i, j, w = line.split()
            i, j, w = int(i), int(j), float(w)  # Parse each element appropriately
            adjacency_list[i].append((j, w))
            adjacency_list[j].append((i, w))  # Since it's undirected
    
    return num_vertices, adjacency_list
```

```python
import heapq  # We'll use Python's heapq to manage the priority queue functionality

class MinHeap:
    def __init__(self):
        self.heap = []  # This will store the (key, id) tuples
        self.position_map = {}  # Maps ids to their position in the heap

    def heap_ini(self, keys):
        # Initialize heap with (key, id) tuples
        for i, key in enumerate(keys, start=1):
            heapq.heappush(self.heap, (key, i))
            self.position_map[i] = key  # Store the current key for each id
    
    def min_key(self):
        # Returns the minimum key without deleting it
        return self.heap[0][0] if self.heap else None
    
    def min_id(self):
        # Returns the id with the minimum key without deleting it
        return self.heap[0][1] if self.heap else None

    def delete_min(self):
        # Deletes and returns the id of the element with the minimum key
        _, min_id = heapq.heappop(self.heap)
        self.position_map.pop(min_id, None)
        return min_id

    def in_heap(self, id):
        # Check if the element with the given id is in the heap
        return id in self.position_map

    def key(self, id):
        # Returns the key of the element with the given id
        return self.position_map.get(id, float('inf'))

    def decrease_key(self, id, new_key):
        # Decreases the key of the element with the given id, if the new key is smaller
        if id in self.position_map and new_key < self.position_map[id]:
            self.position_map[id] = new_key
            heapq.heappush(self.heap, (new_key, id))  # Push the updated 
```

```python
def prim_mst(num_vertices, adjacency_list):
    # Initialize the heap and other necessary structures
    heap = MinHeap()
    mst_edges = []  # To store the edges in the minimum spanning tree
    visited = set()  # Track visited vertices
    keys = [float('inf')] * (num_vertices + 1)  # Initialize keys as infinity
    keys[1] = 0  # Start from vertex 1 (root) with key 0
    
    # Initialize the heap with the keys
    heap.heap_ini(keys[1:])
    
    # Prim's Algorithm Loop
    while heap.heap:
        # Get the vertex with the minimum key
        u = heap.delete_min()
        visited.add(u)
        
        # Iterate over adjacent vertices of u
        for v, weight in adjacency_list[u]:
            if v not in visited and weight < heap.key(v):
                # Update the MST with edge (u, v) and the new key
                mst_edges.append((u, v, weight))
                heap.decrease_key(v, weight)
    
    return mst_edges
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to draw the MST
def draw_mst(mst_edges):
    G = nx.Graph()
    
    # Add edges to the graph
    for u, v, weight in mst_edges:
        G.add_edge(u, v, weight=weight)
    
    # Draw the graph
    pos = nx.spring_layout(G)  # Positions for all nodes
    edge_labels = {(u, v): f"{weight:.2f}" for u, v, weight in mst_edges}  # Edge labels with weights

    plt.figure(figsize=(10, 8))
    nx.draw(G, pos, with_labels=True, node_size=500, node_color="lightblue", font_size=10, font_weight="bold")
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red")
    plt.title("Minimum Spanning Tree (MST)")
    plt.show()

# Call the draw function
draw_mst(mst_edges)
```
